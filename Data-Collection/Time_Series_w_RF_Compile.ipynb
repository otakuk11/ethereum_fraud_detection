```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
df = pd.read_csv('../Data/new_ts_features_only.csv')
```

```python
Y = df[['FLAG']]
```

```python
Y
```

```python
X = df.drop(columns=['FLAG','Unnamed: 0','Address'])
```

```python
X
```

```python
X = X.fillna(0)
```

```python
m = (X == -np.inf)
X2 = X.loc[m.any(axis=1), m.any(axis=0)]
```

```python
X2
```

```python
X = X.drop(columns=['adjusted_eth_value__sample_entropy','adjusted_eth_value__augmented_dickey_fuller__attr_"teststat"__autolag_"AIC"'])
```

```python
X = X.dropna(axis=1, how='all')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=0)
```

```python
# Create a random forest classifier
clf = RandomForestClassifier(n_estimators=10000, random_state=0, n_jobs=-1,verbose=1)

# Train the classifier
clf.fit(X_train, y_train)
```

```python
feat_labels = X.columns.tolist()
```

```python
feat_labels
```

```python
# Print the name and gini importance of each feature
feature_imprt_dict={}
for feature in zip(feat_labels, clf.feature_importances_):
    feature_imprt_dict[feature[0]] = feature[1]
```

```python
Feature_imprt_sorted = {k: v for k, v in sorted(feature_imprt_dict.items(), key=lambda item: item[1],reverse=True)}
```

```python
Feature_imprt_sorted
```

```python
feature_list = list(Feature_imprt_sorted.keys())
  
```

```python
top_10_features = feature_list[:10]
```

```python
df_orig = pd.read_csv('../Data/address_data_combined.csv')
```

```python
top_10_features.insert(0,'Address')
```

```python
df_new = df[top_10_features]
```

```python
final_df = pd.merge(df_orig,df_new,on='Address',how='left')
```

```python
final_df = final_df.dropna()
```

```python
final_df.to_csv('../Data/address_data_combined_ts.csv')
```
